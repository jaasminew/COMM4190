```python
client = Anthropic()
```

## Using the vision models

* https://docs.anthropic.com/claude/docs/vision

* Image data should be in `base64` encoding and added to the `content` value in the `messages`

```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
display(Image('automatic_writing_images/MSColl412_fldr230_body0008.jpg', width=400))
```

```python
encoded_img = encode_image('automatic_writing_images/MSColl412_fldr230_body0008.jpg')
```

```python
message = client.messages.create(
    model="claude-3-5-sonnet-20241022",
    max_tokens=1024,
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image",
                    "source": {
                        "type": "base64",
                        "media_type": "image/jpeg",
                        "data": encoded_img,
                    },
                },
                {
                    "type": "text",
                    "text": "What's the most common character in the manuscript?"
                }
            ],
        }
    ],
)
```
